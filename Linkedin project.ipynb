```python
#import necessary modules
from selenium import webdriver
import time
import pandas as pd
import os

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
#https://www.linkedin.com/jobs/search?keywords=Data%20Analyst&location=Bengaluru%2C%20Karnataka%2C%20India&geoId=105214831&trk=public_jobs_jobs-search-bar_search-submit&redirect=false&position=1&pageNum=0
url1 = 'https://www.linkedin.com/jobs/collections'
```

```python
#driver code

from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
email = "mailabhiraj@gmail.com"
password = "@Bhi1996"

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.linkedin.com/login')
driver.maximize_window()

driver.find_element(By.ID,'username').send_keys(email)
driver.find_element(By.ID,'password').send_keys(password)
driver.find_element(By.ID,'password').send_keys(Keys.RETURN)

driver.get(url1)
driver.implicitly_wait(10)
```

```python
#count number of jobs

n=driver.find_element(By.CLASS_NAME,'results-context-header__job-count')[0].text
```

```python
#removing the '+' from number of jobs

n = ''.join(i for i in n if i.isdigit())
```

```python
#convert to numeric

y=pd.to_numeric(n)
```

```python
#scroll till bottom of page and if 'See More Jobs' is present, click on that and load more jobs.

i=2
while i<=16:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    i=i+1
    
    try:
        x=driver.find_element(By.XPATH,"//button[@aria-label='Load more results']")
        driver.execute_script("arguments[0].click();",x)
        time.sleep(3)
        
    except:
        pass
        time.sleep(2)
```

```python
#empty list to store all company names

companyname=[]
```

```python
#get all company names from specific class name

for j in range(y):
    try:
        company=driver.find_elements(By.CLASS_NAME,'base-search-card__subtitle')[j].text
        companyname.append(company)
    
    except IndexError:
        print("done")
```

```python
#empty list to store all titles of jobs

titlename=[]
```

```python
#get all titles of jobs

for j in range(y):
    try:
        title=driver.find_elements(By.CLASS_NAME,'base-search-card__title')[j].text
        titlename.append(title)
    
    except IndexError:
        print("done")
```

```python
#empty list to store all locations

locname1=[]
```

```python
#get all locations of jobs

for j in range(y):
    try:
        loc=driver.find_elements(By.CLASS_NAME,'job-search-card__location')[j].text
        locname1.append(loc)
    
    except IndexError:
        print("done")
```

```python
cnt=[]
```

```python
#num-applicants__icon num-applicants__icon--notify-pebble lazy-loaded onerror

for j in range(y):
    try:
        count=driver.find_elements(By.CLASS_NAME,'num-applicants__caption')[j].text
        cnt.append(count)
    
    except IndexError:
        print("done")
```

```python
#convert lists to dataframe

companyfinal=pd.DataFrame(companyname,columns=["company"])
```

```python
titlefinal=pd.DataFrame(titlename,columns=["title"])
```

```python
locfinal=pd.DataFrame(locname1,columns=["location"])
```

```python
#join dataframes

f=companyfinal.join(titlefinal)
```

```python
final=f.join(locfinal)
```

```python
#write to csv

final.to_csv('updatedlinkedjobdetails.csv')
```
